```python

def get_defensive_matchups(player_id):
    matchups = LeagueSeasonMatchups(season_type_playoffs='Playoffs', def_player_id_nullable=player_id, off_team_id_nullable=miami_id)
    def_df = matchups.get_data_frames()[0]
    return def_df
```

```python
brown_df = get_defensive_matchups(brown_id)
```

```python
target_ids = [brown_id, kemba_id, tatum_id, smart_id]
player_dfs = []
for cur_id in target_ids:
    player_dfs.append(get_defensive_matchups(cur_id))
```

```python
from nba_api.stats.endpoints.leaguedashteamptshot import LeagueDashTeamPtShot
```

```python
from nba_api.stats.library.parameters import ShotClockRangeNullable
```

```python
boston_id = find_teams_by_full_name("Boston Celtics")
```

```python

def get_shotclock_stats(game_date):
    game_dfs = []
    for shot_clock_range in ["24-22", "22-18 Very Early", "18-15 Early"]:
        cur_game_df = LeagueDashTeamPtShot(season_type_all_star="Playoffs", date_from_nullable=game_date, date_to_nullable=game_date, shot_clock_range_nullable=shot_clock_range).get_data_frames()[0]
        cur_game_df = cur_game_df.loc[:, ['TEAM_NAME', 'FGM', 'FGA']]
        cur_game_df = cur_game_df.loc[cur_game_df.TEAM_NAME == 'Boston Celtics']
        cur_game_df = cur_game_df.rename(columns={'FGM': f'{shot_clock_range}_FGM', 'FGA': f'{shot_clock_range}_FGA'})
        cur_game_df = cur_game_df.reset_index(drop=True)
        game_dfs.append(cur_game_df)
        sleep(0.3)
    return game_dfs
```

```python
game_1_stats = pd.concat(get_shotclock_stats(game_date="09-15-2020"), axis=1)
game_1_stats.loc[:, 'Game'] = 'Game 1'
game_2_stats = pd.concat(get_shotclock_stats(game_date="09-17-2020"), axis=1)
game_2_stats.loc[:, 'Game'] = 'Game 2'
game_3_stats = pd.concat(get_shotclock_stats(game_date="09-19-2020"), axis=1)
game_3_stats.loc[:, 'Game'] = 'Game 3'
```

```python
all_games = pd.concat([game_1_stats, game_2_stats, game_3_stats])
```
